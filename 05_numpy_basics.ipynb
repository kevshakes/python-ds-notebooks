```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Array:", arr)
print("Mean:", arr.mean())
print("Std:", arr.std())
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)
print("column sums:", matrix.sum(axis=0))
print("row sums:", matrix.sum(axis=1))
```

```python
# Practice task:
# Generate 100 random numbers and compute min, max, mean.

# Generate 100 random numbers between 1 and 100
random_numbers = np.random.randint(1, 101, 100)

# Compute min, max, and mean
min_value = random_numbers.min()
max_value = random_numbers.max()
mean_value = random_numbers.mean()

# Print the results
print(f"Min: {min_value}")
print(f"Max: {max_value}")
print(f"Mean: {mean_value:.2f}")
```
