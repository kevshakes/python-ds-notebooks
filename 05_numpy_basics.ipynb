```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
print("Mean:", arr.mean())
print("Std:", arr.std())
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)
print("Column sums:", matrix.sum(axis=0))
print("Row sums:", matrix.sum(axis=1))
```

```python
# Practice task:
import random

# Generate 100 random numbers (between 0 and 1)
numbers = [random.random() for _ in range(100)]

# Compute statistics
minimum = min(numbers)
maximum = max(numbers)
mean = sum(numbers) / len(numbers)

print("Min:", minimum)
print("Max:", maximum)
print("Mean:", mean)
```
