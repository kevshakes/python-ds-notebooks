```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
print("Mean:", arr.mean())
print("Std:", arr.std())
```

```python
import random

# Generate 100 random numbers between 0 and 100
numbers = [random.uniform(0, 100) for _ in range(100)]  # Use randint(0,100) for integers

# Compute statistics
minimum = min(numbers)
maximum = max(numbers)
mean = sum(numbers) / len(numbers)

print("Min:", minimum)
print("Max:", maximum)
print("Mean:", mean)
```
