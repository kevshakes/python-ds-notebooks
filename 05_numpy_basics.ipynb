```python
import numpy as np 
arr = np.array([1, 2, 3, 4, 5])
print('array:', arr)
print('array shape:', arr.shape)
print('array size:', arr.size)
```

```python
import numpy as np

# Create NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Print results
print("Array:", arr)
print("Mean:", arr.mean())
print("Std:", arr.std())
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)
print("Column sums:", matrix.sum(axis=0))
print("Row sums:", matrix.sum(axis=1))
```

```python
import numpy as np

# Generate 100 random numbers between 0 and 1
numbers = np.random.rand(100)

# Compute statistics
minimum = numbers.min()
maximum = numbers.max()
mean_value = numbers.mean()

# Print results
print("Minimum:", minimum)
print("Maximum:", maximum)
print("Mean:", mean_value)
```
