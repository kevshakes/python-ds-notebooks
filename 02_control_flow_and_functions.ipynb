```python
score = 83

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "D"

print("Grade:", grade)
```

```python
# Loops
nums = [2, 4, 6, 8]
total = 0
for n in nums:
    total += n

print("Total:", total)
```

```python
# Functions
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(celsius_to_fahrenheit(25))
```

```python
# Practice task:
# Write a function called is_even(n) that returns True/False
def is_even(n):
    """
    Return True if `n` is even, False otherwise.
    Works for integers and integer-like values.
    """
    try:
        return int(n) % 2 == 0
    except (TypeError, ValueError):
        raise TypeError("is_even expects an integer-like value")

if __name__ == "__main__":
    tests = [0, 1, 2, 3, 4, -1, 2.0]
    for t in tests:
        print(t, is_even(t))
```
