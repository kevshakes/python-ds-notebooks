```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set theme
sns.set_theme(style="whitegrid")

# Create DataFrame
sales = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "revenue": [1200, 1350, 1280, 1500, 1650, 1720]
})

# Create plot
plt.figure(figsize=(8, 4))
plt.plot(sales["month"], sales["revenue"], marker="o")

# Add labels and title
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.tight_layout()
plt.show()
```

```python
students = pd.DataFrame({
    "group": ["A", "A", "B", "B", "C", "C"],
    "score": [70, 75, 80, 85, 90, 95]
})

plt.figure(figsize=(6, 4))
sns.boxplot(data=students, x="group", y="score")
plt.title("Score Distribution by Group")
plt.show()
```

```python
# Practice task:
# Create a histogram of the score column and explain what it shows.
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame
data = pd.DataFrame({
    "score": [55, 70, 75, 80, 90, 65, 88, 92, 73, 60, 77, 85, 69, 95]
})

# Create histogram
plt.figure()
plt.hist(data["score"], bins=5)
plt.title("Distribution of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
