```python
# Practice task:
# Create a histogram of the score column and explain what it shows.
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame
data = pd.DataFrame({
    "score": [55, 70, 75, 80, 90, 65, 88, 92, 73, 60, 77, 85, 69, 95]
})

# Create histogram
plt.figure()
plt.hist(data["score"], bins=5)
plt.title("Distribution of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
